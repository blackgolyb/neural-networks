```python
import numpy as np
import plotly.graph_objects as go
from plotly.offline import init_notebook_mode, iplot

from nexum.core.models import Perceptron
from nexum.core.layers import OutputLayer
from nexum.services.utils import get_clustered_space_image, get_scatters_clustered

init_notebook_mode(connected=True)
```

```python
o_l = OutputLayer(1, activation_function="ReLu")
nn = Perceptron(layers_config=[2, o_l])

input_data = np.array(
    [
        [1, 1],
        [1, -1],
        [2, 0.5],
        [2, -2],
    ]
)

targets = np.array(
    [
        [1],
        [1],
        [0],
        [0],
    ]
)

nn.train(input_data, targets, epochs=50, learning_rate=0.1)


image = get_clustered_space_image(
    nn=nn,
    input_data=input_data,
)

traces = get_scatters_clustered(
    nn=nn,
    input_data=input_data,
)

layout = go.Layout()
fig = go.Figure([image, *traces], layout=layout)
iplot(fig)
```

```python
o_l = OutputLayer(1, activation_function="ReLu")
nn = Perceptron(layers_config=[3, o_l])

input_data = np.array(
    [
        (-1.5, -1.5, -1.5),
        (1.5, -1.5, -1.5),
        (-1.5, 1.5, -1.5),
        (1.5, 1.5, -1.5),
        (-1.5, -1.5, 1.5),
        (1.5, -1.5, 1.5),
        (-1.5, 1.5, 1.5),
        (1.5, 1.5, 1.5),
    ]
)

targets = np.array(
    [
        [1],
        [0],
        [0],
        [0],
        [1],
        [1],
        [1],
        [0],
    ]
)

nn.train(input_data, targets, epochs=50, learning_rate=0.1)

traces = get_scatters_clustered(
    nn=nn,
    input_data=input_data,
)

layout = go.Layout()
fig = go.Figure(traces, layout=layout)
iplot(fig)
```
